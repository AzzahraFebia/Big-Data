### 3. Praktik PySpark Lanjutan
- **Latihan 1**: Memanipulasi Data dengan PySpark.

````python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
````

- **Tugas 2**: Lakukan filter, penghitungan rata-rata, dan pengurutan data menggunakan PySpark.

````python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [
    ("Ali", 34, "Pilot", "Mendengarkan musik", "L"),
    ("Budi", 23, "Mahasiswa", "Renang", "L"),
    ("Citra", 29, "Perawat", "Membaca", "P"),
    ("Dina", 45, "Dokter", "Menonton Film", "P"),
    ("Winatra", 20, "Mahasiswa", "Dance", "P")

]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]

df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
print("DATAFRAME")
df.show()

# Filtering data berdasarkan gender
print("FILTERING DATA (GENDER)")
df_filtered = df.filter(df['Gender'] == "P")
df_filtered.show()

# Menghitung rata-rata usia
print("RATA-RATA USIA")
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data pekerjaan secara ascending
print("URUTAN DATA PEKERJAAN")
df_sorted = df.orderBy("Pekerjaan", ascending=True)
df_sorted.show()
````

### 5. Menggabungkan PySpark dan Pandas

- **Tugas 5**: Gunakan metode ini untuk menggabungkan data yang Anda buat di PySpark dengan data dari Pandas, kemudian lakukan analisis sederhana seperti menghitung rata-rata usia.

````python
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.functions import avg

# Inisialisasi Spark
spark = SparkSession.builder.appName("Konversi").getOrCreate()

# Membuat Data Spark
data_spark = [
    ("Ali", 34, "Pilot", "Mendengarkan musik", "L"),
    ("Budi", 23, "Mahasiswa", "Renang", "L"),
    ("Citra", 29, "Perawat", "Membaca", "P"),
    ("Dina", 45, "Dokter", "Menonton Film", "P"),
    ("Winatra", 20, "Mahasiswa", "Dance", "P")
]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df = spark.createDataFrame(data_spark, columns)

# Membuat Data Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina", "Winatra"],
    "Usia": [34, 23, 29, 45,20]
}
df_pandas = pd.DataFrame(data_pandas)

# Rename usia di Pandas
df_pandas = df_pandas.rename(columns={"Usia": "Usia_Pandas"})

# Konversi Pandas -> PySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

# Menampilkan DataFrame hasil konversi
print("Hasil konversi Pandas -> Spark")
df_spark_from_pandas.show()

# Menggabungkan
df_joined_spark = df.join(df_spark_from_pandas, on="Nama")
print("Gabungan Pandas dan Spark")
df_joined_spark.show()

# Menghitung rata-rata usia
print("Rata-rata usia")
df_joined_spark.select(avg("Usia")).show()
````

- **Tugas 6**: Gabungkan data dari PySpark dan Pandas, lalu lakukan operasi statistik seperti menghitung nilai maksimum usia.

````python
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.functions import max

# Inisialisasi Spark
spark = SparkSession.builder.appName("Konversi").getOrCreate()

# Membuat Data Spark
data_spark = [
    ("Ali", 34, "Pilot", "Mendengarkan musik", "L"),
    ("Budi", 23, "Mahasiswa", "Renang", "L"),
    ("Citra", 29, "Perawat", "Membaca", "P"),
    ("Dina", 45, "Dokter", "Menonton Film", "P"),
    ("Winatra", 20, "Mahasiswa", "Dance", "P")
]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df = spark.createDataFrame(data_spark, columns)

# Membuat Data Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina", "Winatra"],
    "Usia": [34, 23, 29, 45, 20]
}
df_pandas = pd.DataFrame(data_pandas)

# Rename usia di Pandas
df_pandas = df_pandas.rename(columns={"Usia": "Usia_Pandas"})

# Konversi Pandas -> PySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

# Menampilkan Dat=aFrame hasil konversi
print("Hasil konversi Pandas -> Spark")
df_spark_from_pandas.show()

# Menggabungkan
df_joined_spark = df.join(df_spark_from_pandas, on="Nama")
print("Gabungan Pandas dan Spark")
df_joined_spark.show()

# Menghitung nilai maksimum usia
print("Nilai maksimum Usia :")
df_joined_spark.select(max("Usia")).show()
````
