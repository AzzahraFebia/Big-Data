```python
# TUGAS 2

from pymongo import MongoClient

# Koneksi ke MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["company_db"]
collection = db["employees"]

# Hapus data lama biar gak dobel tiap run
collection.delete_many({})

# INSERT DATA 
employee1 = {'Nama': 'Najeya', 'Departemen': 'IT', 'Usia': 27, 'Gaji': 5500}
employee2 = {'Nama': 'Sean', 'Departemen': 'HR', 'Usia': 30, 'Gaji': 5000}
employee3 = {'Nama': 'Joana', 'Departemen': 'IT', 'Usia': 28, 'Gaji': 4500}

# INSERT SATU PER SATU
collection.insert_many([employee1, employee2, employee3])
print("Insert Data Berhasil!\n")

# READ DATA
print("üìã DATA KARYAWAN :")
for emp in collection.find():
    print(f"- {emp['Nama']} | Dept: {emp['Departemen']} | Usia: {emp['Usia']} | Gaji: {emp['Gaji']}")
print("-" * 45)

# UPDATE DATA
collection.update_one({'Nama': 'Sean'}, {'$inc': {'Gaji': 1000}})
print("\n--> Gaji Sean naik 1000!")

collection.update_many({'Departemen': 'IT'}, {'$inc': {'Gaji': 100}})
print("--> Semua gaji di Departemen IT naik 100!\n")

# READ DATA SETELAH UPDATE
print("üìã DATA SETELAH UPDATE :")
for emp in collection.find():
    print(f"- {emp['Nama']} | Dept: {emp['Departemen']} | Usia: {emp['Usia']} | Gaji: {emp['Gaji']}")
print("-" * 45)

# DELETE DATA
collection.delete_one({'Nama': 'Joana'})
print("\n--> Joana dihapus dari daftar karyawan.\n")

# READ DATA SETELAH DELETE
print("üìã DATA SETELAH DELETE :")
for emp in collection.find():
    print(f"- {emp['Nama']} | Dept: {emp['Departemen']} | Usia: {emp['Usia']} | Gaji: {emp['Gaji']}")
print("-" * 45)
```

```python
# INSERT DATA TAMBAHAN

new_employee_IT = {
    'Nama': 'Hanata',
    'Departemen': 'IT',
    'Usia': 26,
    'Gaji': 4800
}

new_employee_HR = {
    'Nama': 'Ken',
    'Departemen': 'HR',
    'Usia': 32,
    'Gaji': 5200
}

collection.insert_many([new_employee_IT, new_employee_HR])
print("Karyawan baru untuk tiap departemen berhasil ditambahkan!\n")

# Tampilkan data setelah insert
print("üìã DATA SETELAH TAMBAH KARYAWAN :")
for emp in collection.find():
    print(f"- {emp['Nama']} | Dept: {emp['Departemen']} | Usia: {emp['Usia']} | Gaji: {emp['Gaji']}")
print("-" * 45)
```

```python
# TUGAS 3
# Hitung rata-rata gaji per departemen
pipeline = [
    {'$group': {'_id': '$Departemen','Rata_rata_Gaji': {'$avg': '$Gaji'}}}
]

print("\n--- RATA-RATA GAJI PER DEPARTEMEN ---")
for hasil in collection.aggregate(pipeline):
    print(f"- Departemen {hasil['_id']}: {hasil['Rata_rata_Gaji']}")
```

```python
# TUGAS 4
# 4.1 Buat Koleksi Produk berisi Data Produk

products = db["products"]

# Hapus data lama
products.delete_many({})

# Tambah data produk (minimal 10)
data_produk = [
    {'Nama': 'Laptop', 'Kategori': 'Elektronik', 'Harga': 9000000},
    {'Nama': 'Kulkas', 'Kategori': 'Elektronik', 'Harga': 4500000},
    {'Nama': 'Smartphone', 'Kategori': 'Elektronik', 'Harga': 6000000},
    {'Nama': 'Printer', 'Kategori': 'Elektronik', 'Harga': 1500000},
    {'Nama': 'Meja Rias', 'Kategori': 'Furniture', 'Harga': 1200000},
    {'Nama': 'Kursi Gaming', 'Kategori': 'Furniture', 'Harga': 2000000},
    {'Nama': 'Lemari Kaca', 'Kategori': 'Furniture', 'Harga': 2500000},
    {'Nama': 'Buku Tulis', 'Kategori': 'ATK', 'Harga': 3000},
    {'Nama': 'Pulpen', 'Kategori': 'ATK', 'Harga': 2500},
    {'Nama': 'Tempat Pensil', 'Kategori': 'ATK', 'Harga': 10000}
]

products.insert_many(data_produk)
print("\nüìã DATA PRODUK")
for produk in data_produk:
    print(f"- {produk['Nama']} | Kategori: {produk['Kategori']} | Harga: {produk['Harga']}")
print("-" * 55)

# 4.2 Query untuk Menemukan Produk dengan Harga > rata-rata
pipeline_Rata_rata= [
    {'$group': {'_id': None, 'Rata_Harga': {'$avg': '$Harga'}}}
]
rata2 = list(products.aggregate(pipeline_Rata_rata))[0]['Rata_Harga']

print(f"\n>> Rata-rata harga produk: {rata2}")
print("\nüìã Produk dengan harga di atas rata-rata")

for p in products.find({'Harga': {'$gt': rata2}}):
    print(f"- {p['Nama']} ({p['Kategori']}) : {p['Harga']}")
print("-" * 45)

# 4.3 Buat Aggregation Pipeline untuk Menghitung Total Produk per-Kategori
pipeline_Total = [
    {'$group': {'_id': '$Kategori', 'Total_Produk': {'$sum': 1}}}
]

print("\nüìã Total Produk per Kategori")
for hasil in products.aggregate(pipeline_Total):
    print(f"- {hasil['_id']}: {hasil['Total_Produk']} produk")
print("-" * 45)
```

```python
# TUGAS 5
# 5.1 Cari 5 Laryawan dengan Gaji Tertinggi setiap Departemen

pipeline_top5 = [
    {'$sort': {'Departemen': 1, 'Gaji': -1}}, {'$group': {'_id': '$Departemen','Top5': {'$push': {'Nama': '$Nama', 'Gaji': '$Gaji'}},}},
    {'$project': {'Top5': {'$slice': ['$Top5', 5]}}}
]

print("\nüìã 5 Karyawan dengan Gaji Tertinggi per Departemen:")
for dept in collection.aggregate(pipeline_top5):
    print(f"\nDepartemen: {dept['_id']}")
    for k in dept['Top5']:
        print(f"- {k['Nama']} : {k['Gaji']}")
```

```python
# 5.2 Hapus Karyawan Berusia < 25 Tahun

hapus = collection.delete_many({'Usia': {'$lt': 25}})
print(f"\nüóëÔ∏è {hapus.deleted_count} karyawan berusia di bawah 25 tahun berhasil dihapus!\n")

print("üìã Data Karyawan Terbaru")
for emp in collection.find():
    print(f"- {emp['Nama']} | Dept: {emp['Departemen']} | Usia: {emp['Usia']} | Gaji: {emp['Gaji']}")
```

```python
# 5.3 Laporan Ringkas (Total Gaji Karyawan per Departemen & Rata-Rata Umur Karyawan)

pipeline_laporan = [
    {'$group': {'_id': '$Departemen','Total_Gaji': {'$sum': '$Gaji'},'Rata_Umur': {'$avg': '$Usia'}}}
]

print("\nüìã Laporan Total Gaji & Rata-rata Umur per Departemen:")
for hasil in collection.aggregate(pipeline_laporan):
    print(f"- {hasil['_id']} | Total Gaji: {hasil['Total_Gaji']} | Rata-rata Umur: {hasil['Rata_Umur']}")
```
